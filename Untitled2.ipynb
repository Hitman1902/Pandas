```python
import pandas as pd
import folium
df=pd.read_csv('California_Houses.csv')
df=df.head(200)
df
lat=36.7783
lon=-119.4179
world_maps=folium.Map(location=[lat,lon],zoom_start=12)
Population=folium.map.FeatureGroup()
for lat,lon in zip(df.Latitude,df.Longitude):
    Population.add_child(folium.features.CircleMarker([lat,lon],radius=5,color='black',fill=True,fill_color='blue'))
world_maps.add_child(Population)
world_maps
```

```python
latitude=list(df.Latitude)
longitude=list(df.Longitude)
label=list(df.Population)
for lat,long,label in zip(latitude,longitude,label):
    folium.Marker([lat,long],popup=label).add_to(world_map)
world_map.add_child(Population)
```
